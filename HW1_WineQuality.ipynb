```python
"""Reading the dataset"""
ds_train=pd.read_csv("/content/wine-quality-white-and-red.csv")
```

```python
"""change the categorical feature type into numerical(we dont use it but there was an error for calculating median,mean etc, so I changed it to not have the annoying error."""
ds_train["type"] = ds_train["type"].map({"white" : 0, "red" : 1})
```

```python
"""Here we can see the mean value, standard deviation, as well as other important informaation such as min and max values which represent the range of values for each collumn"""
ds_train.describe()
```

```python
"""We can see the mode"""
ds_train.mode()
```

```python
"""Such is the median"""
ds_train.median()
```

```python
"""The variance"""
ds_train.var()
```

```python
"""And this is the correlation to the target"""
true_cor=ds_train.corr()["quality"]
corr=abs(ds_train.corr()["quality"]).sort_values(ascending=False)
corr.update(true_cor)
corr.drop(["quality"])
```
